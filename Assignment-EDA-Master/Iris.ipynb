## Seaborn
This graphical library allows to draw all the previously described plots, but it has some additional interesting features that may become handy in data analytics. The following part focuses only on those extra plots.

Basically, seaborn provides a collection of functions thro
ugh different modules: relational, distributional, categorical. For a detailed list of those functions please refer to the [official documentation](https://seaborn.pydata.org/tutorial/function_overview.html).

We'll go through the most interesting functions that were not previously presented in matplotlib.

```python
import pandas as pd # pd alias Pandas library, useful for data manipulation and analysis
import numpy as np # np alias NumPy library, useful for numerical computations
import matplotlib.pyplot as plt # plt alias Matplotlib library, useful for plots

# First way
filename = 'iris.csv' # csv file to load in the current folder
col_names = ['sepal-lenght', 'sepal-widht', 'petal-lenght', 'petal-width', 'class']
dmds = pd.read_csv(filename, names=col_names) # This line reads the CSV file into a DataFrame (a table-like structure in Pandas) and stores it in the variable dmds
data = dmds.copy() # This creates a copy of the original DataFrame (dmds) and assigns it to the variable data.
data.head()
```

```python
# Second way
import seaborn as sns # pd alias Seaborn library, useful for data manipulation and analysis
```

```python
iris = sns.load_dataset('iris') # loads the famous Iris dataset (included directly in Seaborn)
print(iris['species'].unique()) # prints the list of unique species found in the species column. unique() returns distinct values.

iris.head() # It displays the first 5 rows of the DataFrame by default
iris.tail() # It displays the last 5 rows of the DataFrame by default
```

The histogram can be used to represent the distribution of datapoints for a given feature:

```python
%matplotlib inline 
#Jupyter notebook magic command it tells Jupyter to display all Matplotlib plots directly inside the notebook, right below the code cell.
#Do not use this line in a normal Python script — only in notebooks.

plt.figure() # prepares a space where the plot will be drawn
sns.histplot(data=iris, x="sepal_width", hue="species") #plot the distribution of the sepal_width column

# hue="species" → color the bars by flower species
# multiple="stack" → stack the histograms on top of each other instead of overlaying them
```

```python
%matplotlib inline
plt.figure()
sns.histplot(data=iris, x="sepal_length", hue="species", multiple="stack")
```

The distribution of datapoints can also be displayed as separate classes, such as:

```python
sns.histplot(data=iris, x="sepal_length", hue="species")
```

```python
sns.displot(data=iris, x="sepal_width", hue="species", col="species"); # Creates one subplot per species, arranged side by side
```

In order to assess the relationship between two variables a scatter plot may be used, eventually characterizing data by hue and/or size of the markers.

```python
fig, axs = plt.subplots(1, 1, figsize=(8, 8)) #Creates a Matplotlib figure with 1 row and 1 column a single subplot. figsize=(8, 8) sets the figure size to 8×8 inches.
petal_area = iris.petal_length * iris.petal_width
petal_area.rename('Area',inplace=True)

g = sns.scatterplot(data=iris, x="sepal_width", y='sepal_length',hue="species", size=petal_area, ax=axs)
axs.set_xlabel("Sepal Width")
axs.set_ylabel("Sepal Length")
```

Another interesting plot available in seaborn is <code>jointplot</code> which combines the scatter plot with features distribution by class and produces the following output:

```python
sns.jointplot(data=iris, x="sepal_width", y='sepal_length', hue="species");
```

And the best way to visualize relationships among the pairs of features is through the <code>pairplot</code>. The diagonal represents the distribution of values for the selected feature and the off-diagonal plots display the pairwise scatter plot against the other features:

```python
sns.pairplot(data=iris, hue="species");
```

A violin plot plays a similar role as a box but adds the distribution of quantitative data across categorical variables such that those distributions can be compared.

Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

```python
sns.violinplot(data=iris, x='species', y='sepal_length', hue='species');
```
